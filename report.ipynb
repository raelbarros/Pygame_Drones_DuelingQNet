```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# extrair valores de X e Y de uma string
def extrair_valores(s):
  if not pd.isnull(s):
    indice_inicio_x = s.find("x=") + 2
    indice_fim_x = s.find(", y=")

    indice_inicio_y = s.find("y=") + 2
    indice_fim_y = s.find(")")

    if indice_inicio_x != -1 and indice_fim_x != -1 and indice_inicio_y != -1 and indice_fim_y != -1:
      valor_x = float(s[indice_inicio_x:indice_fim_x])
      valor_y = float(s[indice_inicio_y:indice_fim_y])

      return pd.Series({'X': valor_x, 'Y': valor_y})

    # Se a string for nula ou os índices não forem válidos, retorna None
  return pd.Series({'X': None, 'Y': None})


def plot_grafico_colisao(df):
  # Plotar as coordenadas X e Y
  plt.scatter(df['X'], df['Y'])
  # plt.xticks([])
  # plt.yticks([])
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('Plot de Coordenadas X e Y')
  plt.show()
```

# 1 Drone

```python
df_01 = pd.read_csv('/content/(1)2023-11-25 11:05:51.830977.csv')
```

```python
# Aplicar a função ao DataFrame
df_01[['X', 'Y']] = df_01['collision_point'].apply(extrair_valores)

# Exibir o DataFrame resultante
print(df_01[['X', 'Y']].head())
```

```python
plot_grafico_colisao(df_01)
```
